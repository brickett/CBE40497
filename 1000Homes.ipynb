# Predicting Impact of South Bend's Housing Program

During the 1,000 Homes in 1000 Days initiative, the City of South Bend demolished or repaired over 1,000 homes that had been left behind by absentee landlords. The effects were not equally distributed. This code combines data on the initiative with Census demographic data from the American Community Survey to analyze the profile of the areas affected by the initiative.

## Import & Clean Data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os

mydir = os.getcwd()
parcels = pd.read_csv(os.path.join(mydir,'Abandoned_Property_Parcels.csv'))

zips = [46637, 46628, 46635, 46616, 46617, 46615, 46601, 46619, 46613, 46614] # 46618, 46623, 46625 are included in data, but don't exist
zips = sorted(zips)
sbdata = np.zeros((10,9))
sbdata[:,0] = zips

asr = pd.read_csv(os.path.join(mydir,'ACS_15_5YR_DP05.csv')) #age, sex, race
asr = asr.drop([0])
asr = asr.sort_values(['GEO.id2'])
med = pd.read_csv(os.path.join(mydir,'ACS_15_5YR_DP02.csv')) #education, disability
med = med.drop([0])
med = med.sort_values(['GEO.id2'])
um = pd.read_csv(os.path.join(mydir,'ACS_15_5YR_DP03.csv')) #unemployment, median family income
um = um.drop([0])
um = um.sort_values(['GEO.id2'])
```

## Structure Data

```python
#how many houses were torn down in each zip code (try groupby zipcode)
for k in range(0,len(parcels)):
    if round(parcels['Zip_Code'][k],5)*1000 in zips:
        sbdata[zips.index(round(parcels['Zip_Code'][k]*1000,5)),1] += 1

i=1

for k in range(0,len(sbdata)):
    while round(sbdata[k,0]) != float(asr['GEO.id2'][i]) and i<len(asr.index):
        i += 1
    sbdata[k,2] = float(asr['HC03_VC05'][i]) # %female
    sbdata[k,3] = float(asr['HC03_VC29'][i]) # %65+
    sbdata[k,4] = float(asr['HC03_VC49'][i]) # %white

i=1
for k in range(0,len(sbdata)):
    while round(sbdata[k,0]) != float(med['GEO.id2'][i]) and i<len(med.index):
        i += 1
    sbdata[k,5] = float(med['HC03_VC96'][i]) # %bachelor's degree or higher
    sbdata[k,6] = float(med['HC03_VC106'][i]) # %civilian noninstitutionalized population with disability

i=1
for k in range(0,len(sbdata)):
    while round(sbdata[k,0]) != float(um['GEO.id2'][i]) and i<len(med.index):
        i += 1
    sbdata[k,7] = float(um['HC03_VC07'][i]) # %unemployed
    sbdata[k,8] = float(um['HC01_VC114'][i]) # $median family income
```

```python
sbdataf = pd.DataFrame(sbdata, index=zips)
sbdataf = sbdataf.drop([0],1)
sbdataf = sbdataf.rename(index=int, columns={1:'# houses', 2:'%fem', 3:'%65+', 4:'%white', 5:'%bach', 6:'%disability', 7:'%unemployed', 8:'median_fam_income'})
sbdataf
```

## Regression

```python
import statsmodels.api as sm

y = sbdataf['# houses'].values
sbreg = sbdataf.drop(['# houses'],1)
X = sbreg.values
X = sm.add_constant(X)

model1=sm.OLS(y,X)
results1=model1.fit()
results1.summary()
```

## Principal Component Analysis

### Examine Raw Data

Adapted from jckantor [Getting Started with PCA](https://github.com/jckantor/CBE40497/blob/master/Getting%20Started%20with%20PCA.ipynb)

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

%matplotlib inline

# plot distribution of data
plt.figure(figsize=(16,6))
plt.subplot(2,4,1)
k = 1
for c in sbdataf.columns:
    plt.subplot(2,4,k)
    plt.hist(sbdataf[c],normed=True,bins=5)
    plt.title(c)
    k += 1
plt.tight_layout()
```

### Scale Data

Adjust data to be represented by standard deviation rather than raw value.

```python
sbdataf_std = StandardScaler().fit_transform(sbdataf)

lbls = sbdataf.columns

plt.figure(figsize=(16,6))
plt.subplot(2,4,1)
k = 0
for c in lbls:
    plt.subplot(2,4,k+1)
    plt.hist(sbdataf_std[:,k],normed=True)
    plt.title(c)
    k += 1
plt.tight_layout()
```

```python
pca = PCA(n_components=3)
Y = pca.fit_transform(sbdataf_std)

w = pca.components_
v = pca.explained_variance_ratio_
print(v)

for k in range(0,len(w)):
    plt.subplot(3,1,k+1)
    plt.bar(range(0,len(w[k])),w[k],width=.5)
    plt.xticks(range(0,len(w[k])),lbls)
    plt.title('explained variance ratio = {0:.3f}'.format(v[k]))

plt.tight_layout()
```

### Examine for Cluster

```python
plt.scatter(Y[:,0],Y[:,1])
k = 0
for n in zips:
    if sbdataf_std[k,0] < 0:
        plt.text(Y[k,0],Y[k,1],n,color='green')
    elif sbdataf_std[k,0] > 1:
        plt.text(Y[k,0],Y[k,1],n, color='red')
    else:
        plt.text(Y[k,0],Y[k,1],n, color = 'orange')
    k += 1
```
